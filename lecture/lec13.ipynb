## Comparison

There are 6 basic comparison operators:

* Greater Than: >
* Greater Than or Equal To: >=
* Less Than: <
* Less Than or Equal To: <=
* Equal To: ==
* Not Equal To: !=

To use these, you must have two operands on either side of one of the operators above. In addition, there are 3 logical operators:

* AND: All parts must be true for the whole to be true
* OR: One part must be true for the whole to be true
* NOT: Means the opposite; False becomes True and True becomes False

## Simulation

### Use Conditional Statements and Random Selection to Create Simulations


Let's play a game: we each roll a die. 

If my number is bigger: you pay me a dollar.

If they're the same: we do nothing.

If your number is bigger: I pay you a dollar.

Steps:
1. Find a way to simulate two dice rolls.
2. Compute how much money we win/lose based on the result.
3. Do steps 1 and 2 10,000 times.

### Conditional Statements

```python
# Work in progress
def one_round(my_roll, your_roll):
    if my_roll > your_roll:
        return 1
```

```python
one_round(4, 3)
```

```python
one_round(2, 6)
```

```python
# Final correct version
def one_round(my_roll, your_roll):
    if my_roll > your_roll:
        return 1
    elif my_roll < your_roll:
        return -1
    elif my_roll == your_roll:
        return 0
```

```python
one_round(1, 1)
```

```python
one_round(6, 5)
```

```python
one_round(7, -1)
```

```python
# Final correct version
def one_round(my_roll, your_roll):
    faces = np.arange(1,7)
    if sum(my_roll == faces) != 1 or sum(your_roll == faces) != 1:
        print('Please provide a number between 1 and 6')
    elif my_roll > your_roll:
        return 1
    elif my_roll < your_roll:
        return -1
    elif my_roll == your_roll:
        return 0

one_round(7, -1)
```

### Simulating the roll of a die

```python
die_faces = np.arange(1, 7)
```

```python
np.random.choice(die_faces)
```

```python
# Use the simulation to run one round of the game above.
def simulate_one_round():
    my_roll = np.random.choice(die_faces)
    your_roll = np.random.choice(die_faces)
    return one_round(my_roll, your_roll)
```

```python
simulate_one_round()
```

### Use For Statements to Run Simulations Many Times

```python
##Create an loop to store the results of the game played several times in array.
game_outcomes = make_array()

for i in np.arange(5):
    game_outcomes = np.append(game_outcomes, simulate_one_round())
    
game_outcomes
```

```python
game_outcomes = make_array()

for i in np.arange(10000):
    game_outcomes = np.append(game_outcomes, simulate_one_round())
    
game_outcomes
```

```python
#Determine how many outcomes are in the array.
len(game_outcomes)
```

```python
#Display the results as a Table
results = Table().with_column('My winnings', game_outcomes)
results
```

```python
#Display the data of the table as a horizontal bar chart.
results.group('My winnings').barh('My winnings')
```
