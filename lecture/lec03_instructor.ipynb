## Identifer Names

### Rules

1. May only use alphabet or numerical characters
1. May use the underscore _ character
1. First character must be a letter or underscore
1. Should try to ensure name has meaning to the expression or value
1. Identifer name is first, then equal sign to assign an expression or value

### Example

* identifier = "value"
* x = 1
* sample = 1 / 2
* snake_case = "Identifer name uses underscore to separate words. Preferred"
